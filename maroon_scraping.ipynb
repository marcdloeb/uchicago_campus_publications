```python
def makemydir(folder, file_parent):
  try:
    os.makedirs(file_parent + folder)
  except OSError:
    pass
  # let exception propagate if we just can't
  # cd into the specified directory
```

```python
def get_issues(file_parent, txt_parent, year, page_link):
    """
    Takes a campus publications page, and extracts the 20 links that appear on it, saves the pdfs
    """
    issues_links_list = []
    result = requests.get(page_link)
    doc = BeautifulSoup(result.text, "html.parser")

    #finding all the individual Maroon editions on the page
    issue_root = "https://campub.lib.uchicago.edu"
    issues = doc.find_all("div", class_="span8")
    print("number of issues:", len(issues))

    if not issues:
        print("no issues, something is wrong,reading doc again")
        issues = doc.find_all("div", class_="span8")
        print("number of issues:", len(issues))

    for issue in issues:
        
        issue_page = issue.find("a")["href"]
        issue_tail = issue_page.split("/")[2]
        issue_tag = "_".join(issue_tail.split("-")[-2:])
        print(issue_tag)

        #issue_link = issue_root + issue_page
        issue_pdf_link = issue_root + "/pdf/" + issue_tail
        issue_file = file_parent + year + "/" + issue_tag + ".pdf"
        issue_txt_location = txt_parent + year + "_ocr/ocr_" + issue_tag + ".txt"
        #print(issue_link)
        print(issue_pdf_link)
        print(issue_file)

        issues_links_list += [(issue_pdf_link, issue_txt_location, issue_file, issue_tag)]
        
        #issue_result = requests.get(issue_pdf_link)
        #print(issue_result.headers.get('content-type'))
        #open(issue_file, 'wb').write(issue_result.content)                 # turn these back on

    return issues_links_list
```

```python
issues_df = pd.DataFrame(columns=["year", "link", "txt", "pdf", "tag"])
links_and_paths = {}
file_parent = "web_scraping/pdfs/"
txt_parent = "txt/"

#getting all the years from the main page, nested loop below finds the extra pages for each year
for year_link in year_links:
    print("year_link:", year_link)
    
    #making the directory
    year = year_link[-4:]
    print("year is", year)
    #makemydir(year, file_parent) #####turn this back on

    #opening the year link
    result = requests.get(year_link)
    year_doc = BeautifulSoup(result.text, "html.parser")
   
    #saving the pdfs from the parent page
    issues_links_list = get_issues(file_parent, txt_parent, year, year_link)    #REMEMBER!!!! this last one has to be year_link, not extra_page_link
    links_and_paths[year] = issues_links_list
    year_df = pd.DataFrame(issues_links_list, columns=["link", "txt", "pdf", "tag"])
    year_df["year"] = year
    issues_df = pd.concat([issues_df, year_df])

    #finding extra pages
    #finding p-tags with "page" in them, which indicates a next page box (otherwise not given its own tag or class)
    root = year_link + "&startDoc="
    page_tags = [tag.text for tag in year_doc.find_all("p") if 'Page' in tag.text]
    #finding the largest number, creating links for the additional pages
    if page_tags:
        for extra_page_link in [root + str(num * 20 + 1) for num in range(1, max([int(s) for s in re.findall(r'\d+', (page_tags[0]))]))]:
            print("year is", year)
            print("extra_page:", extra_page_link)
            #getting the pdfs from the children pages
            issues_links_list = get_issues(file_parent, txt_parent, year, extra_page_link)
            links_and_paths[year] = issues_links_list
            year_df = pd.DataFrame(issues_links_list, columns=["link", "txt", "pdf", "tag"])
            year_df["year"] = year
            issues_df = pd.concat([issues_df, year_df])
            
            print(len(links_and_paths[year]))
    print("------------------------------------")
```
